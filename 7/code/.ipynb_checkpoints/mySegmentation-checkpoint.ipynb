```python
smooth_baboon=cv2.GaussianBlur(baboon_array,(3,3),1)
```

```python
baboon_array_gray=baboon_array[:,:,0]*0.3+baboon_array[:,:,1]*0.59+baboon_array[:,:,2]*0.11
```

```python
def create_5_data(img_arr):
    r=img_arr.shape[0]*img_arr.shape[1]
    mean_data=np.zeros((img_arr.shape[0]*img_arr.shape[1],5))
    data=np.zeros((img_arr.shape[0]*img_arr.shape[1],5))
    num=0
    for i in range(img_arr.shape[0]):
        for j in range(img_arr.shape[1]):
            data[num,:3]=img_arr[i,j]
            data[num,3]=i
            data[num,4]=j
            num+=1
    return data,mean_data
```

```python
data,mean_data=create_5_data(smooth_baboon[::2,::2])
```

```python
data_c=np.copy(data)
```

```python
def fnc_gaussian(x,sigma):
    return np.exp(- (np.linalg.norm(x,2) ** 4) / (2 * sigma ** 2))
```

```python
for k in range(10):
    print(k)
    nbrs = NearestNeighbors(n_neighbors=100,algorithm='auto').fit(data_c)
    dis,ind=nbrs.kneighbors(data_c)
    for i in range(data.shape[0]):
        #print(ind[i].shape)
        #print(i)
        sum_d=np.zeros((1,5),dtype='float64')
        sum_w=0
        for j in (ind[i]):
            w_r=fnc_gaussian(data[j,:3]-data[i,:3],4)
            w_s=fnc_gaussian(data[j,3:5]-data[j,3:5],8)
            sum_d[0,:3]+=data[j,:3]*w_r*w_s
            sum_w+=w_r*w_s
        data_c[i,:3]=sum_d[0,:3]/sum_w
```

```python
img_op=np.zeros_like(baboon_array[::2,::2])
for d in data_c:
    img_op[int(d[3]),int(d[4])]=d[:3]
```

```python
plt.imshow(img_op)
plt.imsave("op20.png",img_op)
plt.imsave("original.png",smooth_baboon[::2,::2])
```
